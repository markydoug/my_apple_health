### Deal with time zones. It will take a long time.

Going through my photos for the time period I have Apple health data for. I will separate dataframes it into periods  of time, then convert the timezone of those times while I was in China.

```python
records_list['creation_time'] = pd.to_datetime(records_list['creation_time'])
records_list['activity_start_time'] = pd.to_datetime(records_list['activity_start_time'])
records_list['activity_end_time'] = pd.to_datetime(records_list['activity_end_time'])
```

```python
records_list = records_list.set_index('creation_time').sort_index()
```

```python
def change_time_zone(df):
    #convert index timezone
    df = df.tz_convert('Asia/Chongqing')
    #convert activity_start_time timezone
    df['activity_start_time'] = pd.to_datetime(df['activity_start_time']).dt.tz_convert('Asia/Chongqing')
    #convert activity_end_time timezone
    df['activity_end_time'] = pd.to_datetime(df['activity_end_time']).dt.tz_convert('Asia/Chongqing')
    df['in_china'] = 1
    return df
```

```python
def add_date(df):
    df['date'] = df['activity_start_time'].dt.strftime('%Y/%m/%d')
    return df
```

```python
def add_times(df):
    df['start'] = df['activity_start_time'].dt.strftime('%H:%M')
    df['end'] = df['activity_end_time'].dt.strftime('%H:%M')
    return df
```
