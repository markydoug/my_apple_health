### Steps

```python
step_counts = records_list[records_list['activity_type'] == 'StepCount']
```

```python
step_counts.head()
```

```python
step_counts.info()
```

```python
step_counts['source'].value_counts()
```

```python
step_counts['value'] = step_counts['value'].astype(int)
```

```python
step_counts.info()
```

```python
step_counts = step_counts.set_index('date')
```

```python
step_counts
```

```python
daily_steps = step_counts.resample('D')[['value']].sum()
```

```python
#look at all my steps per day
daily_steps.plot()
```

```python
daily_steps['2020-05'].plot()
```

```python
xiao_mi_steps = step_counts[step_counts['source'] == '小米运动']
```

```python
xiao_mi_steps.sort_index()
```

**Definitely an issue here**

```python
not_xiao_mi = step_counts[step_counts['source'] != '小米运动']
```

```python
not_xiao_mi.resample('D')[['value']].sum().plot()
```

```python
phone_steps = step_counts[step_counts['source'] == 'Marky Doug']
```

```python
phone_steps.resample('D')[['value']].sum().plot()
```

```python
apple_watch_steps = step_counts[(step_counts['source'] != 'Marky Doug') & (step_counts['source'] != '小米运动')]
```

```python
plt.figure(figsize = (20,8))
plt.plot(phone_steps.loc['2021-11':].resample('D')[['value']].sum(), label='Phone', linewidth=1)
plt.plot(apple_watch_steps.resample('D')[['value']].sum(), label='Watch', linewidth=1)
plt.legend()
plt.show()
```

Looks like my apple watch will collect more steps. I will go with the watch step counts for the time when I have a watch and the phone for when I didn't have the watch.

```python
steps = phone_steps.loc[:'2021-10-28'].append(apple_watch_steps)
```

```python
steps.resample('D')[['value']].sum().plot()
```

```python
daily_step_count = steps.resample('D')[['value']].sum()
```

What day did I have the most steps?

```python
daily_step_count[daily_step_count.value.max() == daily_step_count.value]
```

What day did I have the fewest steps?

```python
daily_step_count[daily_step_count.value.min() == daily_step_count.value]
```
